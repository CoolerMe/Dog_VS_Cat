```python
#先加载所有的train数据
from sklearn.datasets import load_files
from keras.utils import np_utils
import numpy as np
import os

#加载所有的数组,此处应该有更优雅的实现的
def load_data(path):
    file_names=os.listdir(path)
    file_targets=[]
    for file_name in file_names:
        if 'dog' in file_name:
            file_targets.append(1)
        else:
            file_targets.append(0)
    data_files=np.array(file_names)
    data_targets=np.array(file_targets)
    return data_files,data_targets

data_files,data_targets=load_data('./train')
print(data_files[0])
```

```python
#使用ResNet50来判断照片类型
from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np

model = ResNet50(weights='imagenet')

img_path = os.path.join('./train','cat.0.jpg')
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

preds = model.predict(x)
# decode the results into a list of tuples (class, description, probability)
# (one such list for each sample in the batch)
result=decode_predictions(preds, top=3)[0]
print('Predicted:', result)
```

```python
%%timeit
#top n的准确率
top_n=10
print(len(data_files))
for i in range(len(data_files)):
    file_name =data_files[i]
    img_path = os.path.join('./train',file_name)
    img = image.load_img(img_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    
    preds = model.predict(x)
    result=decode_predictions(preds, top=top_n)[0]
    #先当它是有问题的
    has_problem = True
    #如果当前目标是狗
    if data_targets[i]:
        for j in range(top_n):
            if result[j][0]  in dogs:
                has_problem=False
                break
    else:
        for j in range(top_n):
            if result[j][0]  in cats:
                has_problem=False
                break
    if has_problem:
        print('This picture has problem:',file_name)
    
    if i>=100:
        break;
```
